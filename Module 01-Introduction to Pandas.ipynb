# Numeric DataType in Pandas

```python
import  pandas as pd
```

```python
type(data)
```

```python
df = pd.DataFrame(data)
df
    
```

```python
df.info()
```

```python
df.dtypes
```

```python
#creating data type to int32
df['age'] = df['age'].astype('int32')
df['salary'] = df['salary'].astype('int32')
df
```
